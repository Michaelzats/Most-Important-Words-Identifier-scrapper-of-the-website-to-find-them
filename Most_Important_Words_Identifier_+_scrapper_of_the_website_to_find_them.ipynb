```python
# Most important words from the website

import requests
from bs4 import BeautifulSoup
from sklearn.feature_extraction.text import TfidfVectorizer

# Specify the URL of the website you want to scrape
url = 'https://de.deskbird.com/'

# Make a request to the website
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the paragraph tags
paragraphs = soup.find_all('p')

# Extract the text from the paragraph tags
text = [p.text for p in paragraphs]

# Join the texts into a single string
text = ' '.join(text)

# Create a TfidfVectorizer object with the specified stop words
vectorizer = TfidfVectorizer(stop_words=["die","eine","um","vor,""euer","oder","über", "das","ist","war","sind", "der", "deine", "deines", "deinem", "ihnen", "und","zu", "für", "an", "in", "im", "mit","ein","von","vom","unsere","einen","den","des","innen","wir","deinen","dein","du","the","with","your","and","of","you","without","we","you","they","to","will","by","it","our","unserer"])

# Fit the vectorizer on the text
vectorizer.fit([text])

# Get the feature names and values
feature_names = vectorizer.get_feature_names()
feature_values = vectorizer.transform([text]).toarray()

# Find the most important words
important_words = sorted(zip(feature_values[0], feature_names), reverse=True)[:10]

# Print the most important words
print(important_words)
```
